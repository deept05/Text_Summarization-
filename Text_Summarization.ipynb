```python
import bs4 as bs
import urllib.request
import re

scraped_data = urllib.request.urlopen("https://en.wikipedia.org/wiki/Machine_learning")
article = scraped_data.read()

parsing = bs.BeautifulSoup(article,'lxml')
paragraph = parsing.find_all('p')

text_article = ""
for p in paragraph:
  text_article += p.text
```

```python
text_article
```

```python
text_article = re.sub(r'\[0-9*\]', ' ', text_article)
text_article = re.sub(r'\s+', ' ', text_article)
```

```python
formatted_text_article = re.sub('[^a-zA-Z]', ' ', text_article)
formatted_text_article = re.sub(r'\s+', ' ', formatted_text_article).strip()
```

```python
import nltk
nltk.download('punkt')
sentence_list = nltk.sent_tokenize(text_article)
stopwords = nltk.corpus.stopwords.words('english')
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}
for word in nltk.word_tokenize(formatted_text_article):
  if word not in stopwords:
    if word not in word_frequencies.keys():
      word_frequencies[word] = 1
    else:
      word_frequencies[word] += 1




# word_frequencies = {
#     'Machine': 1,
#     'learning': 1,
#     'ML': 2,  # "ML" appears twice
#     'field': 1,
#     'AI': 1,
#     'algorithms': 1,
#     'learn': 1,
#     'data': 1,
#     'used': 1,
#     'many': 1,
#     'applications': 1,
#     'healthcare': 1,
#     'finance': 1,
#     'more': 1,
#     'systems': 1,
#     'supervised': 1,
#     'unsupervised': 1,
#     'reinforced': 1
# }
```

```python
maximun_frequency = max(word_frequencies.values())
for word in word_frequencies.keys():
  word_frequencies[word] = (word_frequencies[word]/maximun_frequency)



# word_frequencies = {
#     'Machine': 0.5,         # 1 / 2 = 0.5
#     'learning': 0.5,        # 1 / 2 = 0.5
#     'ML': 1.0,              # 2 / 2 = 1.0 (most frequent word)
#     'field': 0.5,           # 1 / 2 = 0.5
#     'AI': 0.5,              # 1 / 2 = 0.5
#     'algorithms': 0.5,      # 1 / 2 = 0.5
#     'learn': 0.5,           # 1 / 2 = 0.5
#     'data': 0.5,            # 1 / 2 = 0.5
#     'used': 0.5,            # 1 / 2 = 0.5
#     'many': 0.5,            # 1 / 2 = 0.5
#     'applications': 0.5,    # 1 / 2 = 0.5
#     'healthcare': 0.5,      # 1 / 2 = 0.5
#     'finance': 0.5,         # 1 / 2 = 0.5
#     'more': 0.5,            # 1 / 2 = 0.5
#     'systems': 0.5,         # 1 / 2 = 0.5
#     'supervised': 0.5,      # 1 / 2 = 0.5
#     'unsupervised': 0.5,    # 1 / 2 = 0.5
#     'reinforced': 0.5       # 1 / 2 = 0.5
# }
```

```python
sentences_scores = {}
for sent in sentence_list:
  for word in nltk.word_tokenize(sent.lower()):
    if word in word_frequencies.keys():
      if(len(sent.split(' ')) < 30):
        if sent not in sentences_scores.keys():
          sentences_scores[sent] = word_frequencies[word]
        else:
          sentences_scores[sent] += word_frequencies[word]


# sentences_scores = {
#     "Machine learning is fascinating.": 2.5,
#     "Artificial intelligence and machine learning are the future.": 3.5,
#     "The applications of machine learning are numerous.": 3.0
# }
```

```python
import heapq
summary_sentences = heapq.nlargest(5, sentences_scores, key=sentences_scores.get)

summary = ' '.join(summary_sentences)

print(summary)
```
